The observer of a line of sight is at $(0, 0)$ and the target is at $(D_d, 0)$. Find $y$ in terms of $x$, $C$, $D_d$, and $R_\oplus$.

- $y$ represents the vertical distance of a light ray from the straight line between the observer and target.
- $x$ represents the horizontal distance along the straight line between the observer and target.
- $C$ represents the amount of light curvature, $R_\text{Light} = C R_\oplus$.
- $D_d$ represents the distance between the observer and target along the straight line between them, the direct distance.
- $R_\oplus$ represents the average radius of the Earth at sea level.

---
 
Equation of a circle:
$$(C R_\oplus)^2 = \left(x-\frac{D_d}{2}\right)^2 + (y + \ell)^2 , \quad \ell = \sqrt{(C R_\oplus)^2 - \left(\frac{D_d}{2}\right)^2}$$
$$(C R_\oplus)^2 = x^2 - D_d x + \left(\frac{D_d}{2}\right)^2 + y^2 + 2\ell y + (C R_\oplus)^2 - \left(\frac{D_d}{2}\right)^2$$
$$0 = y^2 + 2\ell y + x(x-D_d)$$
Quadratic Formula:
$$y = \frac{-2\ell + \sqrt{(2\ell)^2 - 4x(x-D_d)}}{2}$$
$$y = \sqrt{\ell^2 - x(x-D_d)} - \ell$$
