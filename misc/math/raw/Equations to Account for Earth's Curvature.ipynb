Find $D_x$, $D_y$, and $\alpha$ in terms of $D_s$ and $R_\oplus$. 

- $D_x$ represents the distance between the observer and the target's projection on the observer's horizontal plane.
- $D_y$ represents the vertical distance between the target and the observer's horizontal plane.
- $\alpha$ represents the angle between the observer's horizontal plane and the straight line between the observer and target.
- $D_s$ represents distance between the observer and target along the surface of the Earth at sea level.
- $R_\oplus$ represents the average radius of the Earth at sea level.
- $\theta$ represents the angle between the line from the observer to Earth's centre and the line from the target to Earth's centre. 

---

$$\sin(\theta) = D_x / R_\oplus$$
$$\sin(D_s / R_\oplus) = D_x / R_\oplus$$
$$D_x = R_\oplus \sin(D_s / R_\oplus)$$


$$R_\oplus = R_\oplus \cos(\theta) + D_y$$
$$D_y = R_\oplus - R_\oplus \cdot \cos(D_s / R_\oplus)$$
$$D_y = R_\oplus \left( 1 - \cos(D_s / R_\oplus) \right)$$


$$\tan(\alpha) = D_y / D_x$$
$$\tan(\alpha) = \frac{R_\oplus \left( 1 - \cos(D_s / R_\oplus) \right)}{R_\oplus \sin(D_s / R_\oplus)}$$
$$\alpha = \arctan \left( \frac{1 - \cos(D_s / R_\oplus)}{\sin(D_s / R_\oplus)} \right)$$
$$\alpha = \arctan \left( \tan \left( \frac{D_s}{2 R_\oplus} \right) \right)$$
$$\alpha = \frac{D_s}{2 R_\oplus}, \quad 0 < \alpha < \pi/2$$
