In polar coordinates, the observer of a line of sight is at $P_1 = (R_\oplus + h_1, 0)$ and the target is at $P_2 = (R_\oplus + h_2, D_s/R_\oplus)$. Light follows the arc of a circle with radius $R_L = CR_\oplus$. Find the contrast of the target from the perspective of the observer in terms of $h_1$, $h_2$, $D_s$, $H$, $S$, $a$, $\beta_0$, $C$, and $R_\oplus$.

- $c$ represents the contrast of the target from the perspective of the observer.
- $h_1$ represents the elevation of the observer.
- $h_2$ represents the elevation of the target.
- $D_s$ represents the arc distance between the observer and target at sea level.
- $H$ represents the scale height of the atmosphere. 
- $S$ represents the ratio of the light path that is in shade. The shaded region is contiguous and starts at the observer.
- $a$ represents the ratio of irradiation in the shaded region compared to the unshaded region.
- $\beta_0$ represents the scattering coefficient at sea level.
- $\beta_1$ represents the average scattering coefficient along the shaded portion of the light path.
- $\beta_2$ represents the average scattering coefficient along the unshaded portion of the light path.
- $R_\oplus$ represents the average radius of the Earth at sea level.
- $x$ represents the arc distance between the observer and target at sea level.
- $h(x)$ represents the vertical distance of light from sea level at x.
- $L(x)$ represents the distance of from the centre of the Earth to the light arc at x.
- $(M_x, M_y)$ represents the centre of the light arc circle in Cartesian coordinates.
- $D_d$ represents the distance between the observer and target along the straight line between them, the direct distance.

Equation 9, Below the horizon — the physics of extreme visual ranges, Michael Vollmer:

$$c = \frac{e^{-\beta_2 d_2}}{1 - a + \left( \dfrac{a}{e^{-\beta_1 d_1}} \right)}$$

Equation 7, Below the horizon — the physics of extreme visual ranges, Michael Vollmer:

$$\beta_1 = \frac{\beta_0}{d_1}\int_{0}^{d_1} e^{- \left( \tfrac{h(x)}{H} \right)} dx$$
$$\beta_2 = \frac{\beta_0}{d_2}\int_{d_1}^{D_s} e^{- \left( \tfrac{h(x)}{H} \right)} dx$$

$$d_1 = S D_s$$
$$d_2 = (1-S) D_s$$

$$h(x) = L(x) - R_\oplus$$

Equation of the light circle in polar co-ordinates:

$$L(x) = M_x\cos(\theta) + M_y\sin(\theta) + \sqrt{(M_x\cos(\theta) + M_y\sin(\theta))^2 + R_L^2 - M_x^2 - M_y^2}, \quad \theta = \frac{x}{R_\oplus}$$

$$M_x = \frac{x_1 + x_2}{2} - Z\left(\frac{y_2 - y_1}{D_d}\right)$$
$$M_y = \frac{y_1 + y_2}{2} + Z\left(\frac{x_2 - x_1}{D_d}\right)$$

$$Z = \sqrt{R_L^2 - \left(\frac{D_d}{2}\right)^2}$$

$$D_d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$$

$$P_1 = (x_1, y_1), \quad x_1 = R_\oplus + h_1, \quad y_1 = 0$$
$$P_2 = (x_2, y_2), \quad x_2 = (R_\oplus + h_2) \cos(\phi), \quad y_2 = (R_\oplus + h_2) \sin(\phi), \quad \phi = \frac{D_s}{R_\oplus}$$
